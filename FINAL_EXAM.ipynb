def display_menu():
    """
    Display menu options - Week 3 formatting techniques only

    >>> display_menu()  # doctest: +SKIP
    1. Add task
    2. List tasks
    3. Remove task
    4. Quit
    """
    print("1. Add task")
    print("2. List tasks")
    print("3. Remove task")
    print("4. Quit")


def get_user_choice():
    """
    Get and validate user choice - Week 6 input validation style

    >>> # get_user_choice()  # doctest: +SKIP
    """
    while True:
        choice = input("Enter your choice (1-4): ")
        if choice in ["1", "2", "3", "4"]:
            return int(choice)
        else:
            print("Invalid choice. Please enter a number between 1 and 4.")


def add_task(task_list):
    """Add a new task to the task list - Week 4 list manipulation"""
    task = input("Enter the task description: ")
    task_list.append(task)
    print("Task added.")


def list_tasks(task_list):
    """List all tasks - Week 5 for loop"""
    if not task_list:
        print("No tasks found.")
    else:
        for i, task in enumerate(task_list):
            print(f"{i + 1}. {task}")


def remove_task(task_list):
    """Remove a task by its number - Week 6 try/except and validation"""
    if not task_list:
        print("No tasks to remove.")
        return

    list_tasks(task_list)
    try:
        task_num = int(input("Enter the number of the task to remove: "))
        if 1 <= task_num <= len(task_list):
            removed = task_list.pop(task_num - 1)
            print(f"Removed: {removed}")
        else:
            print("Invalid number. No task removed.")
    except:
        print("Please enter a valid number.")


def main():
    """Main programme loop - Week 5 loop patterns only"""
    task_list = []
    while True:
        display_menu()
        choice = get_user_choice()

        if choice == 1:
            add_task(task_list)
        elif choice == 2:
            list_tasks(task_list)
        elif choice == 3:
            remove_task(task_list)
        elif choice == 4:
            print("Goodbye!")
            break


if __name__ == "__main__":
    import doctest
    doctest.testmod()
